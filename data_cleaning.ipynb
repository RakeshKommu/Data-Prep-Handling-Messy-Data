# Task 1 -- Clean the Data

```python
import pandas as pd
import numpy as np
```

```python
# Create DataFrame
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

# ---- Handle Missing Values ----


```python
# Fill age with median
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
# Fill city with mode
df['city'].fillna(df['city'].mode()[0], inplace=True)
```

```python
# ---- Remove Duplicates ----
df.drop_duplicates(inplace=True)
```

```python
print("Cleaned Data:")
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# One-Hot Encoding for 'city'
df = pd.get_dummies(df, columns=['city'])
```

```python
# Label Encoding for 'gender'
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
print("\nEncoded Data:")
print(df)
```

# Task 3 -- Scale Numerical Features

```python
# Copy dataframe for scaling
df_minmax = df.copy()
df_standard = df.copy()
```

```python
# Features to scale
cols = ['age', 'annual_income']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# ---- Min-Max Scaling ----
minmax = MinMaxScaler()
df_minmax[cols] = minmax.fit_transform(df_minmax[cols])
```

```python
print("\nMin-Max Scaled Data:")
print(df_minmax)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# ---- Standard Scaling ----
standard = StandardScaler()
df_standard[cols] = standard.fit_transform(df_standard[cols])
```

```python
print("\nStandard Scaled Data:")
print(df_standard)
```

### Explanation

Min-Max Scaling scales the data between 0 and 1 and is useful when the algorithm requires a bounded range, such as KNN or Neural Networks. Standardization transforms the data to have a mean of 0 and a standard deviation of 1, making it suitable for algorithms like Logistic Regression and SVM. The choice depends on the data distribution and the requirements of the model.
